```r
library(readr)
library(ggplot2)
```

```r
ba <- read_csv("result.csv")
```

```r
spec(ba)
```

```r
ggplot(data = ba, aes(x = NDVI, y = BurntArea))+geom_point()
```

```r
ggplot(data = ba, aes(x = LST, y = BurntArea))+geom_point()
```

```r
df <- dplyr::select_if(ba, is.numeric)

# calulate the correlations
r <- cor(df, use="complete.obs")
round(r,2)
```

```r
install.packages("ggcorrplot")
library(ggplot2)
library(ggcorrplot)
ggcorrplot(r)
```
